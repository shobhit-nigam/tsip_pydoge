```python
from bs4 import BeautifulSoup
```

```python
page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.title)
```

```python
print(soup.title.text)
```

```python
print(soup.body)
```

```python
print(soup.head)
```

```python
if soup.select('h1'):
    print("h1 exists")
```

```python
if soup.select('h2'):
    print("h2 exists")
```

```python
for i in range(1, 7):
    if soup.select('h'+str(i)):
        print('h'+str(i), "exists")
    else:
        print('h'+str(i), "does not exists")
```

```python
print(soup.select('h1'))
```

```python
for val in soup.select('h1'):
    print(val.text)
    print("------")
```

```python
for val in soup.select('h1'):
    print(val.text.strip())
    print("------")
```

```python
h1_tags = []

for val in soup.select('h1'):
    h1_tags.append(val.text)
```

```python
h1_tags
```

```python
images_link = []
```

```python
images = soup.select('img')
```

```python
for val in images:
    src = val.get('src')
    images_link.append(src)
```

```python
for data in images_link:
    print(data)
```

```python
images_link = []

for val in images:
    src = val.get('alt')
    images_link.append(src)
    
for data in images_link:
    print(data)
```

```python
url = "https://web.archive.org/web/20130209050253/http://oreilly.com/store/samplers.html"
```

```python
rg = requests.get(url)
```

```python
rg.status_code
```

```python
soup = BeautifulSoup(rg.content)
```

```python
samples = soup.find_all("a", "item-title")
```

```python
dictd = {}

for val in samples:
    title = val.string.strip()
    dictd[title] = val.attrs['href']
```

```python
for k, v in dictd.items():
    print(k)
    print(v)
    print("-------")
```
